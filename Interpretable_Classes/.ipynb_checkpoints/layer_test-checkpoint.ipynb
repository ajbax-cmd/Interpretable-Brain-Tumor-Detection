# Evaluate Layer Feature Space Correlation with Model Output Using Pearson Correlation
##### -YOLOv8m weights trained on BR35H dataset

```python
import os
import torch
import gc
import matplotlib.pyplot as plt
from yolov8_interpretable import InterpretableYOLOTest

def evaluate_multiple_layers(data_yaml_path, weights_path, layer_indices, pca_components, batch_size=1, img_size=(640, 640)):
    pearson_scores = {}
    for layer_index in layer_indices:
        print(f"Evaluating layer at index: {layer_index}")
        weights = load_model(weights_path, model_class=None)
        model = InterpretableYOLOTest(data_yaml_path, weights, target_layer_index=layer_index, pca_components=pca_components)
        p_score = model.calculate_pearson_correlation() 
        pearson_scores[layer_index] = p_score
        del model
        gc.collect()
    return pearson_scores

def load_model(model_path, model_class=None):
    model = torch.load(model_path)
    model = model['model'].float()
    model.eval()  
    return model

def plot_scores(layer_indices, pearson_scores):
    scores = [pearson_scores[idx] for idx in layer_indices]
    
    # Line plot
    plt.figure(figsize=(12, 6))
    plt.plot(layer_indices, scores, marker='o')
    plt.title('Pearson Scores by Layer Index')
    plt.xlabel('Layer Index')
    plt.ylabel('Pearson Score')
    plt.grid(True)
    plt.show()
    

data_yaml_path = '/home/alan/Documents/Dataset_1/brain-tumor-detection-dataset/Br35H-Mask-RCNN/data.yaml'
weights_path = '/home/alan/Documents/YOLOV8_interpretable/YOLOv8_weights/best.pt'


# Evaluate layers at every 20th index and add specific layers
layer_indices = list(range(140, 295, 10))  # every 10th layer
n_components = list(range(10, 75, 5))
specific_layers = [292, 293, 294]
layer_indices.extend(specific_layers)
layer_indices = sorted(set(layer_indices))  # Ensure no duplicates and sort
pca_pscore = {} #dict {n_component : (layer_index, pearson_correlation) }

for component in n_components:
    pearson_scores = evaluate_multiple_layers(data_yaml_path, weights_path, layer_indices, component)
    print("Layer Scores:")
    for layer_index in layer_indices:
        print(f"Layer {layer_index} - Pearson Score: {pearson_scores[layer_index]}")
    
    # Choose the best layer based on the scores
    best_pearson_index = max(pearson_scores, key=pearson_scores.get)
    print(f"Best pearson layer index: {best_pearson_index} with score: {pearson_scores[best_pearson_index]}")
    
    plot_scores(layer_indices, pearson_scores)

    pca_pscore[component] = (best_pearson_index, pearson_scores[best_pearson_index])

print(pca_pscore)
highest_score = 0
highest_index = 0
for component in pca_pscore.keys():
    if (pca_pscore[component][1] > highest_score):
        highest_score = pca_pscore[component][1]
        highest_index = component

print("PCA component with highest pearson score: " + str(highest_index) + " Most correlated layer: " + str(pca_pscore[highest_index][0]) + " pearson score: " + str(pca_pscore[highest_index][1]))
            


```

```python
keys = list(pca_pscore.keys())
values = list(pca_pscore.values())
layers = [value[0] for value in values]
p_scores = [value[1] for value in values]


plt.bar(keys, p_scores, tick_label=keys, width=4, color='blue')
for i in range(len(keys)):
    plt.text(keys[i], p_scores[i], layers[i], ha = 'center')
plt.xlabel('n Components - PCA Dimension Reduction')
plt.ylabel('Pearson Correlation')
plt.title('YOLOv8 Layer with Highest Pearson Correlation Over n Components')
plt.show()
```
